```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X=np.random.randn(100,3)
for i in range(X.shape[1]):
    X[:,i].sort()
X
```

```python
y= 2.0*X[:,2].reshape(100,1) +1.0 + 0.5*np.random.randn(100,1)
y
```

```python
a=np.ones_like(X[:,0]).reshape((X.shape[0],1))
X=np.hstack((a,X))  # asa facem x_0=1
X
```

```python
teta=np.random.randn(X.shape[1])
teta
```

```python
def h(teta,X):
    return teta.T.dot(X)
```

```python
def J(teta,X,y):
    a=(1/(2*len(y)))
    s=0
    for i in range(1,len(y)):
        s+=(h(teta,X[i,:])-y[i])**2
    return a*s
```

```python

ni=0
c=J(teta,X,y)
a=np.array([ni,c])
a
```
