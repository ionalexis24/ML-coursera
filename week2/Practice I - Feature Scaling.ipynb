Sa zicem ca am un set de date de genul $X=\{x_1,...,x_n\}$ cu $x_i=[x_i^{(j)}\mid j=1,m]$ si fiecare $x_i^{(j)}\in [a_i^{(j)},b_i^{(j)}]$

In principiu vrem ca $x_i^{(j)}\in [-1,1]$. In principiu pentru ca daca intervalul nu e farte diferit ca lungime de $[-1,1]$ nu e nevoie de rescalare. Dar daca intervalul de valori e foarte mare sau foarte mic o rescalare ar fi buna.

Mean normalization.

$x_i=\frac{x_i-\mu_i}{s_i}$ pentru $i>0$ pentru ca $x_0=1$

$\mu_i=\text{mean}(x_i)$ si $s_i=\text{range}(x_i)$

$\text{mean}(x_i)=\frac{1}{m}\sum_j^m x_i^{(j)}$ si $\text{range}(x_i)= \text{max}(x_i)-\text{min}(x_i)$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X=np.random.randn(100,3)
for i in range(X.shape[1]):
    X[:,i].sort()
X  # sa zicem ca asta e un set de date
```

```python
y= 2.0*X[:,2].reshape(100,1) +1.0 + 0.5*np.random.randn(100,1)
y
```

```python
XX=X[:,2]
yy=y
```

```python
X[:,2].mean()
```

```python
X[:,2].max()-X[:,2].min()
```

Normalizam cu medie si cu SD

```python
for i in range(X.shape[1]):
    a=(X[:,i]-X[:,i].mean())
    b=X[:,i].max()-X[:,i].min()
    X[:,i]=a/b
X
```

```python
a=(y-y.mean())
b=y.max()-y.min()
y=a/b
y
```

```python

plt.figure(figsize=(10,5))
plt.subplot(121)
plt.plot(XX,yy)
plt.subplot(122)
plt.plot(X[:,2],y)
plt.show()  #sunt exact la fel lol numai ca axele sunt scalate lol
```
