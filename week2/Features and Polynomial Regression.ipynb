```python
def h(teta,X):
    return teta.T.dot(X)
```

```python
def J(teta,X,y):
    a=(1/(2*len(y)))
    s=0
    for i in range(1,len(y)):
        s+=(h(teta,X[i,:])-y[i])**2
    return a*s
```
