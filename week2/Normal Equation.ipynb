```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X=np.random.randn(100,1)
for i in range(X.shape[1]):
    X[:,i].sort()
X
```

```python
y=-5+2*np.sin(X[:,0].reshape((-1,1)))-3*(X[:,0].reshape((-1,1))**3)+0.2*np.random.randn(100,1)
y
```

```python
X=np.hstack((X,X**2,X**3))
X
```

```python
plt.plot(X[:,0],y,".")
plt.show()
```

```python
def h(teta,X):
    return teta.T.dot(X)
```

```python
a=np.ones_like(X[:,0]).reshape((X.shape[0],1))
X=np.hstack((a,X))  # asa facem x_0=1
X
```

Acum o sa putem sa rezolvam analitic pe teta lol!!!

```python
a=(X.T).dot(y)
b=np.linalg.inv((X.T).dot(X))
teta=b.dot(a)
teta
```

```python
ynn=np.array([])
for i in range(X.shape[0]):
    ynn=np.append(ynn,h(teta,X[i,:]))  #calculam h(teta,X) pentru plot
```

```python
plt.plot(X[:,1],y,".",
        X[:,1],ynn,)
plt.show()
```
